```python
import pandas as pd
import os
```

```python
dataframes = []

for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    df['Artist'] = file.replace('.csv','')
    dataframes.append(df)
```

```python
all_data = all_data[all_data['Artist'] != 'BTS']
all_data.reset_index(drop=True, inplace=True)
print(all_data['Artist'].value_counts())
```

```python
ed_sheeran_lyrics = all_data[all_data['Artist'] == 'EdSheeran']['Lyric']
print(ed_sheeran_lyrics.head(10))
```

```python
lyrics_data = all_data[['Lyric']].copy()
lyrics_data.dropna(subset=['Lyric'], inplace=True)
lyrics_data.reset_index(drop=True, inplace=True)
print(lyrics_data.head(-10))
print("Shape of the dataset:", lyrics_data.shape)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
lyrics_data['Cleaned_Lyrics'] = lyrics_data['Lyric'].apply(preprocess_text)
print(lyrics_data[['Lyric', 'Cleaned_Lyrics']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(lyrics_data['Cleaned_Lyrics'])
print("TF-IDF Matrix Shape:", X.shape)
```

# Using Latent Dirchlet Allocation

```python
from sklearn.decomposition import LatentDirichletAllocation
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42, max_iter=10)
lda.fit(X)
print("Document-Topic Matrix Shape:", lda.transform(X).shape)
```

```python
feature_names = vectorizer.get_feature_names_out()
def print_topics(model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx + 1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[-n_top_words:]]))
        print("-" * 50)
print_topics(lda, feature_names)
```

```python
import numpy as np
document_topics = lda.transform(X)
lyrics_data['Topic'] = np.argmax(document_topics, axis=1)
print(lyrics_data[['Lyric', 'Topic']].head())
```

```python
topic_to_emotion = {
    0: "Introspection",
    1: "Love",
    2: "Happiness",
    3: "Hope",
    4: "Melancholy"
}


lyrics_data['Emotion'] = lyrics_data['Topic'].map(topic_to_emotion)

print(lyrics_data[['Lyric', 'Topic', 'Emotion']].head(10))
```

```python
import matplotlib.pyplot as plt

emotion_counts = lyrics_data['Emotion'].value_counts()
emotion_counts.plot(kind='bar', color='skyblue', figsize=(8, 6))
plt.title('Emotion Distribution Across Lyrics')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Using K means Clustering

```python
from sklearn.cluster import KMeans

n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=48)
kmeans.fit(X)
lyrics_data['Cluster'] = kmeans.labels_
print(lyrics_data[['Lyric', 'Cluster']].head())
```

```python
feature_names = vectorizer.get_feature_names_out()
def print_clusters(kmeans_model, feature_names, n_top_words=20):
    for cluster_idx in range(n_clusters):
        # Sort feature indices by their importance to the cluster
        word_indices = kmeans_model.cluster_centers_[cluster_idx].argsort()[-n_top_words:]
        top_words = [feature_names[i] for i in word_indices]
        print(f"Cluster {cluster_idx}: {', '.join(top_words)}")
        print("-" * 50)
print_clusters(kmeans, feature_names)
```

```python
cluster_to_emotion = {
    0: "Affectionate",
    1: "Anger",
    2: "Melancholy",
    3: "Reflection",
    4: "Love"
}

lyrics_data['Cluster_Emotion'] = lyrics_data['Cluster'].map(cluster_to_emotion)
print(lyrics_data[['Lyric', 'Cluster', 'Cluster_Emotion']].head())
```

```python
cluster_emotion_counts = lyrics_data['Cluster_Emotion'].value_counts()
cluster_emotion_counts.plot(kind='bar', color='skyblue', figsize=(8, 6))
plt.title('Cluster Emotion Distribution Across Lyrics')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Sentiment Analysis

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader_analyzer = SentimentIntensityAnalyzer()

def get_vader_sentiment(text):
    sentiment_scores = vader_analyzer.polarity_scores(text)
    if sentiment_scores['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

lyrics_data['VADER_Sentiment'] = lyrics_data['Lyric'].apply(get_vader_sentiment)

print(lyrics_data[['Lyric', 'VADER_Sentiment']].head())
```
